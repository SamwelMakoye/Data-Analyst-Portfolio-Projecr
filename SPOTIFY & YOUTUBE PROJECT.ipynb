```python
compilation_mean = round(spotify_youtube_data[spotify_youtube_data['Album_type'] == 'compilation']['Comments'].mean(), 0)
compilation_mean
```

```python
#FUNCTION TO FILL THE MISSING DATA WITH RESPECTIVE MEAN VALUES
def impute_mean(cols):
    Comments = cols[0]
    Album_type = cols[1]
    
    if pd.isnull(Comments):
        if Album_type == 'album':
            return album_mean
        elif Album_type == 'single':
            return single_mean
        else:
            return compilation_mean
        
    else:
        return Comments
```

```python
spotify_youtube_data['Comments'] = spotify_youtube_data[['Comments', 'Album_type']].apply(impute_mean, axis = 1)
```

```python
official_stream_mean = round(spotify_youtube_data[spotify_youtube_data['official_video'] == True]['Stream'].mean(), 0)
```

```python
unofficial_stream_mean = round(spotify_youtube_data[spotify_youtube_data['official_video'] == False]['Stream'].mean(), 0)
```

```python
def impute_mean_stream(cols):
    Stream = cols[0]
    official_video= cols[1]
    
    if pd.isnull(Stream):
        if official_video == True :
            return official_stream_mean
        elif official_video == False:
            return unofficial_stream_mean
        
    else:
        return Stream
```

```python
spotify_youtube_data['Stream'] = spotify_youtube_data[['Stream', 'official_video']].apply(impute_mean_stream, axis = 1)
```

```python
def classify_danceability(Danceability):
    for i in range(0, len(spotify_youtube_data) +1):
        if Danceability >= 0 and Danceability <= 0.3:
            return 'A'
        elif Danceability > 0.3 and Danceability <= 0.7:
            return 'B'
        elif Danceability > 0.7:
            return 'C'
```

```python
spotify_youtube_data['Danceability'] = spotify_youtube_data['Danceability'].apply(classify_danceability)
```

```python
spotify_youtube_data['Danceability'].value_counts()
```

```python
def classify_energy(Energy):
    for i in range(0, len(spotify_youtube_data) +1):
        if Energy >= 0 and Energy <= 0.3:
            return 'Low Energy'
        elif Energy > 0.3 and Energy <= 0.7:
            return 'Medium Energy'
        elif Energy > 0.7:
            return 'High Energy'
```

```python
spotify_youtube_data['Energy'] = spotify_youtube_data['Energy'].apply(classify_energy)
```
